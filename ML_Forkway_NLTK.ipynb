```python
# Lemmatization and filtering of skills

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english') + stopwords.words('russian'))

def preprocess_text(text):
    tokens = word_tokenize(text.lower())
    lemmas = [lemmatizer.lemmatize(token) for token in tokens]
    filtered_tokens = [lemma for lemma in lemmas if lemma not in stop_words]

     # We leave only IT terms
    it_skills = [token for token in filtered_tokens if token in it_terms]

    return it_skills
```

```python
# record start time
start = time.time()

# Creating a dictionary and recording candidate ID lists

skills_dict = {}

for index, row in data.iterrows():
    skills = preprocess_text(row['description'])
    for skill in skills:
        if skill not in skills_dict:
            skills_dict[skill] = []
        skills_dict[skill].append(str(row['id']))

# record end time
end = time.time()

# print the difference between start
# and end time in secs
print("The time of execution of above program is :",
      round(end-start, 2), "secs")
```

```python
# Writing a json file

with open('submit.json', 'w', encoding='UTF-8') as f:
  f.write(json.dumps(skills_dict, ensure_ascii=False, indent=4))
```
